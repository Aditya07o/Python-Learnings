## Methods/Functions of list

```python
v1=["apple","mango","orange","cherry"]
print(v1)
```

```python
## methods for addition of element in list:-
## 1. Append() :- we can add single element at a time and that element will bve added at the end of list.
## 2. Extend() :- we can add multiple elements at a time adn the element will be added at the end of list similar to append.
## 3. insert() :- we can add elements at specific position in the list.
```

```python
v1.append("banana")
print(v1)
```

```python
v1.append(["banana","chikku"])
print(v1)
```

```python
## extend
v1.extend(["pineapple","dragon fruit"])
print(v1)
```

```python
## add kiwi fruit at the place of 0th index
v1.insert(0,"kiwi")
print(v1)
```

```python
## methods for removing elements from these list
## 1. pop() :- it will remove last element from the list and if we pass index position inside these pop function
##             then that particular element will be deleted.
```

```python
v1.pop()
print(v1)
```

```python
v1.pop(5)
print(v1)
```

```python
## 2.remove() :- we have to provide element which we want to remove 
v1.remove("kiwi")
print(v1)
```
