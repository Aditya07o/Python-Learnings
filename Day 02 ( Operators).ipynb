## Types of Operators:-

```python
## There are 7 types of operators ionj python
## 1. Logical operator:- AND, OR, NOT
```

```python
## and operator:- if both conditions are true then you will get result otherwise we will get result from else statement 
marks=80
attendance=73
if marks>70 and attendance>75:
  print("pass")
else: 
  print("fail")
```

```python
## or operator:- if any one condition is true then also you will get result  
marks=80
attendance=73
if marks>70 or attendance>75:
  print("pass")
else: 
  print("fail")
```

```python
## not operator:- It is just negation of true result.
marks=80
attendance=73
if not (marks>70 or attendance>75):
  print("fail")
else: 
  print("pass")
```

```python
## Membership operator:- in, not in
L1=["apple", "mango", "pineapple", "orange", "cherry"]
print("cherry" in L1)
```

```python
L1=["apple", "mango", "pineapple", "orange", "cherry"]
print("cherry" not in L1)
```

```python
## Identity operator:- is, is not
T1=(2,4,5,6)
T2=(4,5,6,2)

print(T1 is T2)
```

```python
T1=(2,4,5,6)
T2=(2,4,5,6)

print(T1 is T2)
```

```python
T1=(2,4,5,6)
T2=(2,4,5,6)

print(T1 is not T2)
```
