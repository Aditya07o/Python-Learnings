## Data Structures

```python
## List - [], mutable, ordered, heterogenous, allows duplicate, changable

L1=[30,50.2,True,"Python",30,50.2]
print(L1)
```

```python
L1[2]=150
print(L1)
```

```python
L1.count(50.2)
```

```python
len(L1)
```

```python
## Set - {}, unordered, mutable, not allows duplicate, heterogenous

S1={30,50.2,True,"Python",30,50.2}
type(S1)
```

```python
print(S1)
```

```python
## Tuple- (), ordered, allow duplicate, heterogenous, unchangable(immutable)

T1=(50,"Aditya",True,2.6,50,50)
print(T1)
type(T1)
```

```python
T1[2]=100
```

```python
## Dictionary -{k:v} key value pair madhe, heterogenous, ordered, key cant be dupolicate but values can be, mutable(changable)

D1={"name":"aditya","subject":"python","marks":"92"}
print(D1)
type(D1)
```

```python
D1["marks"]=95
print(D1)
```
